### Function definition

```python
def get_condition_type(row):
    #condition: consioucs, unconscious, break
    #recognition_state: d, post, pre
    #repetition: 1,2,3
    
    # recognized
    #conscious
    if  (row['recognition'] =='y') & (row['condition'] =='conscious') & (row['recognition_state'] =='d'): 
        return 'recognized-con-d'
   
    elif  (row['recognition'] =='y') & (row['condition'] =='conscious') & (row['recognition_state'] =='pre'):
        return 'recognized-con-pre'
   
    elif  (row['recognition'] =='y') & (row['condition'] =='conscious') & (row['recognition_state'] =='post'):
        return 'recognized-con-post'
   
    #unconscious  
    elif  (row['recognition'] =='y') & (row['condition'] =='unconscious') & (row['recognition_state'] =='d'):
        return 'recognized-uncon-d'
   
    elif  (row['recognition'] =='y') & (row['condition'] =='unconscious') & (row['recognition_state'] =='pre'):
        return 'recognized-uncon-pre'
   
    elif  (row['recognition'] =='y') & (row['condition'] =='unconscious') & (row['recognition_state'] =='post'):
        return 'recognized-uncon-post'
       
    #break    
    elif  (row['recognition'] =='y') & (row['condition'] =='break') & (row['recognition_state'] =='d'):
        return 'recognized-break-d'
   
    elif  (row['recognition'] =='y') & (row['condition'] =='break') & (row['recognition_state'] =='pre'):
        return 'recognized-break-pre'
   
    elif  (row['recognition'] =='y') & (row['condition'] =='break') & (row['recognition_state'] =='post'):
        return 'recognized-break-post'
    
    # unrecognized
       #conscious
    if  (row['recognition'] =='n') & (row['condition'] =='conscious') & (row['recognition_state'] =='d'):
        return 'unrecognized-con-d'
   
    elif  (row['recognition'] =='n') & (row['condition'] =='conscious') & (row['recognition_state'] =='pre'):
        return 'unrecognized-con-pre'
   
    elif  (row['recognition'] =='n') & (row['condition'] =='conscious') & (row['recognition_state'] =='post'):
        return 'unrecognized-con-post'
   
    #unconscious  
    elif  (row['recognition'] =='n') & (row['condition'] =='unconscious') & (row['recognition_state'] =='d'):
        return 'unrecognized-uncon-d'
   
    elif  (row['recognition'] =='n') & (row['condition'] =='unconscious') & (row['recognition_state'] =='pre'):
        return 'unrecognized-uncon-pre'
   
    elif  (row['recognition'] =='n') & (row['condition'] =='unconscious') & (row['recognition_state'] =='post'):
        return 'unrecognized-uncon-post'
       
    #break    
    elif  (row['recognition'] =='n') & (row['condition'] =='break') & (row['recognition_state'] =='d'):
        return 'unrecognized-break-d'
   
    elif  (row['recognition'] =='n') & (row['condition'] =='break') & (row['recognition_state'] =='pre'):
        return 'unrecognized-break-pre'
   
    elif  (row['recognition'] =='n') & (row['condition'] =='break') & (row['recognition_state'] =='post'):
        return 'unrecognized-break-post'
    
    
     # no resp
       #conscious
    if  (row['recognition'] =='0') & (row['condition'] =='conscious') & (row['recognition_state'] =='d'):
        return 'noresp-con-d'
   
    elif  (row['recognition'] =='0') & (row['condition'] =='conscious') & (row['recognition_state'] =='pre'):
        return 'noresp-con-pre'
   
    elif  (row['recognition'] =='0') & (row['condition'] =='conscious') & (row['recognition_state'] =='post'):
        return 'noresp-con-post'
   
    #unconscious  
    elif  (row['recognition'] =='0') & (row['condition'] =='unconscious') & (row['recognition_state'] =='d'):
        return 'noresp-uncon-d'
   
    elif  (row['recognition'] =='0') & (row['condition'] =='unconscious') & (row['recognition_state'] =='pre'):
        return 'noresp-uncon-pre'
   
    elif  (row['recognition'] =='0') & (row['condition'] =='unconscious') & (row['recognition_state'] =='post'):
        return 'noresp-uncon-post'
       
    #break    
    elif  (row['recognition'] =='0') & (row['condition'] =='break') & (row['recognition_state'] =='d'):
        return 'noresp-break-d'
   
    elif  (row['recognition'] =='0') & (row['condition'] =='break') & (row['recognition_state'] =='pre'):
        return 'noresp-break-pre'
   
    elif  (row['recognition'] =='0') & (row['condition'] =='break') & (row['recognition_state'] =='post'):
        return 'noresp-break-post'
    else:
        return 'none_recognizable_condition'
```

```python
for subnum in subnum_list:
    print(f'----------sub-{subnum}----------')


    try:
        resp=glob(f'/Volumes/Yunjie/Recognition_experiment//Data/p2_behavioral_data/sub-{subnum}/*response*.csv')
        for file in resp:
            resp_df=pd.read_csv(file)
            break_count= resp_df.break_response.value_counts()

            #if any unconscious trial breaks for the subject
            if len(break_count.index)>1:
                resp_df.loc[(resp_df['break_response'] != '0') & (resp_df['condition'] == 'unconscious'), 'condition'] = 'break'
                print('It BREAKS for the subject!')
            else:
                resp_df.loc[(resp_df['break_response'] != 0) & (resp_df['condition'] == 'unconscious'), 'condition'] = 'break'
                print('no break')

            # sum up for recognition state (pre, post, d)
        grouped = resp_df.groupby('Run')
        grouped_dataframes = {}
        for group_name, group_df in grouped:
            grouped_dataframes[int(group_name)] = group_df

        recognition_state_list=[]
        for i in range(len(grouped_dataframes)):
            disambiguation_list= grouped_dataframes[i+1].loc[(grouped_dataframes[i+1]['image_type'] == 'grayscale') & 
                                        (grouped_dataframes[i+1]['condition'] == 'conscious') & 
                                        (grouped_dataframes[i+1]['repetition']==1), 'image_name'].tolist()
            print(f'for this run{i+1}, the disambiguation images are {disambiguation_list}')

            #"get_trial_type"pre-defined at 0.2.1
            recognition_state_list.append(grouped_dataframes[i+1].apply(get_trial_type, axis=1)) 


        flattened_list = [item for sublist in recognition_state_list for item in sublist]
        resp_df['recognition_state']= flattened_list
        print(resp_df['recognition_state'].unique())

        #"get_condition_type"pre-defined at 0.2.1
        resp_df['trial_type']= resp_df.apply(get_condition_type, axis=1)
        resp_df = resp_df.loc[:, ['onset_time_clock','presentation_duration_clock','path','condition', 'recognition_state','repetition', 'recognition', 'response_time','trial_type','sub_ID','Dominant_eye', 'Run']]
        resp_df.columns= ['onset','duration','path','condition', 'recognition_state','repetition', 'recognition', 'response_time','trial_type','sub_ID','Dominant_eye', 'Run']
        # Get the unique values in the 'run' column
        unique_runs = resp_df['Run'].unique()

        # Loop through each unique run and save it to a separate TSV file
        for run_value in unique_runs:
            # Select the rows with the current run value
            tsv_file_for_run = resp_df[resp_df['Run'] == run_value]
            filename = f"run_{run_value}.tsv"
            directory_path = f'/Volumes/Yunjie/Recognition_experiment/Data/p2_behavioral_response/sub-{subnum}'
            if not os.path.exists(directory_path):
                os.makedirs(directory_path)
            tsv_file_for_run.to_csv(os.path.join(f'/Volumes/Yunjie/Recognition_experiment/Data/p2_behavioral_response/sub-{subnum}',filename), sep='\t', index=False)
            
    except KeyError:
        print("Error, check the subject")
```

## Renaming T1, Functional files

```python
# all subject loop
for subid in subnum_list:
    print(f'Turning file for sub-{subid} into BIDs format')
    
    resp=glob(f'/Volumes/Yunjie/Recognition_experiment/Data/p2_rawdata/sub-{subid}/resp/*response*.csv')
    

    #changing filename for anat
    file_list=glob(f'/Volumes/Yunjie/Recognition_experiment/Data/p2_rawdata/sub-{subid}/anat/*t1*')
    for i in file_list:
        old_name= i
        sub_num=i.split('/')[-3][4:]
        file_postfix=i.split('/')[-1].split('.')[-1]
        new_name= os.path.join(f'/Volumes/Yunjie/Recognition_experiment/Data/p2_rawdata/sub-{sub_num}/anat/',f'sub-{sub_num}_T1w.{file_postfix}')
        os.rename(old_name, new_name)

    #changing filename for func
    file_list=glob(f'/Volumes/Yunjie/Recognition_experiment/Data/p2_rawdata/sub-{subid}/func/*')
    non_tsv_files = [file for file in file_list if not file.endswith('.tsv')]
    tsv_files= [file for file in file_list if file.endswith('.tsv')]
    
    ##change the non-tsv filename
    for i in non_tsv_files:
        old_name= i
        print(old_name)
        sub_num=i.split('/')[-3][4:]
        run_num=int(i.split('/')[-1].split('_')[-1].split('.')[0])-2
        file_postfix=i.split('/')[-1].split('.')[-1]
        new_name= os.path.join(f'/Volumes/Yunjie/Recognition_experiment/Data/p2_rawdata/sub-{sub_num}/func/',f'sub-{sub_num}_task-recognition_run-{run_num}_bold.{file_postfix}')
        os.rename(old_name, new_name)
    
    ##change the tsv filename
    for i in tsv_files:
        old_name= i
        print(old_name)
        sub_num=i.split('/')[-3][4:]
        run_num=i.split('/')[-1].split('_')[-1].split('.')[0]
        file_postfix=i.split('/')[-1].split('.')[-1]
        new_name= os.path.join(f'/Volumes/Yunjie/Recognition_experiment/Data/p2_rawdata/sub-{sub_num}/func/',f'sub-{sub_num}_task-recognition_run-{run_num}_events.{file_postfix}')
        os.rename(old_name, new_name)
```
